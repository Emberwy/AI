```python
import matplotlib
x=[1,2,3,4,5]
y=[2,3,4,5,6]
print(x,y)
```

```python
from matplotlib import pyplot as plt
fig1=plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig2=plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('test2.csv',encoding='utf-8')
print(type(data),data)
```

```python
x=data.loc[:,'x']
print(type(x))
y=data.loc[:,'y']
print(y)
```

```python
# 数据筛选
c=data.loc[:,'x'][y>9]
print(c)
```

```python
# 转化成array
data_array=np.array(data)
print(type(data_array))
print(data_array)
```

```python
# 将数据处理存储起来
data_new=data-10
print(data_new)
data_new.head()
data_new.to_csv('data_new.csv')
```
