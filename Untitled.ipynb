```python
import matplotlib
x=[1,2,3,4,5]
y=[2,3,4,5,6]
print(x,y)
```

```python
from matplotlib import pyplot as plt
fig1=plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig2=plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
